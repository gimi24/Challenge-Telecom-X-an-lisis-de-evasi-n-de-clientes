##**Informe final**

###**Introducci√≥n**

La compa√±√≠a de telecomunicaciones Telecom X enfrenta un elevado √≠ndice de bajas de clientes y requiere identificar las causas detr√°s de este problema para implementar medidas que lo reduzcan. Utilizando los datos proporcionados por la empresa, es necesario realizar procesos de extracci√≥n y depuraci√≥n de datos, junto con un an√°lisis exploratorio, para determinar qu√© variables est√°n influyendo en la significativa p√©rdida de clientes.

###**Limpieza y Tratamiento de Datos**

**Procesamiento de Datos desde la API de Telecom X**  

Los datos se obtienen desde la API de Telecom X en formato JSON, los cuales incluyen informaci√≥n clave sobre los clientes, como datos demogr√°ficos, tipo de servicio contratado y estado de cancelaci√≥n (churn). A continuaci√≥n, se detallan los pasos realizados para su limpieza y preparaci√≥n:  

1. **Detecci√≥n de Inconsistencias**  
   - Se exploraron las columnas del DataFrame para verificar tipos de datos y detectar valores nulos o vac√≠os.  
   - Se extrajeron los diccionarios de las columnas `customer`, `phone`, `internet` y `account` para un an√°lisis completo.  

2. **Manejo de Datos Faltantes**  
   - En la columna **Churn**, se eliminaron registros vac√≠os, ya que representaban menos del 5% de los datos y no afectaban significativamente el an√°lisis.  
   - En la columna **Total**, los valores faltantes se reemplazaron con **0**, dado que correspond√≠an a clientes nuevos que a√∫n no hab√≠an realizado pagos.  
   - Se convirtieron columnas num√©ricas a tipos de dato adecuados (`int` o `float`).  

3. **Creaci√≥n de Nueva Variable**  
   - Se gener√≥ una columna de **"cuentas diarias"**, que calcula el costo diario de los servicios contratados.  

4. **Estandarizaci√≥n de Datos**  
   - Valores categ√≥ricos como *"Yes"/"No"* se transformaron en **1 y 0** para facilitar el an√°lisis.  
   - Se renombraron columnas en espa√±ol para una mejor interpretaci√≥n.  



###**An√°lisis Exploratorio de Datos**

### **An√°lisis Exploratorio de la Evasi√≥n de Clientes en Telecom X**  

Con los datos preparados, se llev√≥ a cabo un an√°lisis exploratorio para identificar patrones y factores asociados a la cancelaci√≥n de clientes. A trav√©s de visualizaciones y m√©tricas estad√≠sticas, se obtuvieron los siguientes hallazgos:  

#### **1. Distribuci√≥n de Cancelaciones (Churn)**  
- De un total de **7,043 clientes**, **1,869 (26.5%)** cancelaron su suscripci√≥n.  
- **Tipo de contrato:**  
  - Los clientes con contrato **mes a mes** presentan la mayor tasa de abandono: de **3,875 usuarios**, **1,655 (42.7%)** dieron de baja su servicio.  
- **M√©todo de pago:**  
  - Los usuarios que pagan mediante **cheque electr√≥nico** son m√°s propensos a cancelar: de **2,365 clientes**, **1,071 (45.3%)** abandonaron el servicio.  

#### **2. An√°lisis de Variables Num√©ricas**  
- **Total de pagos (histograma):**  
  - La mayor√≠a de las cancelaciones se concentran en el rango de **$0 a $285**, lo que sugiere que los clientes tienden a abandonar el servicio cuando los costos son bajos.  
- **Meses de permanencia (boxplot):**  
  - El **primer cuartil** est√° cerca del m√≠nimo, con una **media alrededor de los 10 meses**, lo que indica que muchas cancelaciones ocurren **antes del tercer mes**.  
- **Mensualidad (boxplot):**  
  - Las bajas comienzan a incrementarse cuando el precio supera los **$60**, concentr√°ndose principalmente en el **cuarto cuartil** (clientes con planes m√°s caros).  

### **Conclusiones Clave**  
- Los clientes con contratos **mes a mes** y pagos por **cheque electr√≥nico** presentan mayor riesgo de cancelaci√≥n.  
- La mayor√≠a de las bajas ocurren en los **primeros meses** y en planes con **precios bajos o muy altos**.  



###**Conclusiones**

La mayor p√©rdida de clientes ocurre en usuarios con contrato mes a mes y pago por cheque electr√≥nico, donde la baja comienza a notarse desde el tercer mes. Si el cliente supera los diez meses, es m√°s probable que se quede. Por otro lado, los usuarios con contrato anual o pago autom√°tico tienden a ser m√°s fieles y cancelan menos.

###**Recomendaciones**

**Propuestas para reducir la fuga de clientes:**  

1. **Programa de fidelidad**  
   - Implementar recompensas por permanencia, como descuentos mensuales o premios por antig√ºedad.  

2. **Incentivos para cambio de contrato**  
   - Ofrecer descuentos (ej: primeros 4 meses) a usuarios que migren de planes **mes a mes** a contratos **anuales/bianuales**.  

3. **Beneficios temporales para clientes mes a mes**  
   - Regalar contenido exclusivo (ej: suscripciones de streaming) por tiempo limitado para aumentar engagement.  

4. **Promoci√≥n de pagos autom√°ticos**  
   - Otorgar descuentos a quienes cambien de **cheque electr√≥nico** a m√©todos autom√°ticos (tarjeta/d√©bito).  

5. **Benchmark competitivo**  
   - Realizar un estudio comparativo con otras telecomunicaciones para identificar ventajas/desventajas de Telecom X.
